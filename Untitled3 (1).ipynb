```python
import os
import tarfile
from google.colab import drive
import shutil # Import shutil for rmtree

# Mount drive
drive.mount('/content/drive')

# Using different workspace directories
ARCHIVE_LOC = '/content/drive/MyDrive/HW1'
WORKSPACE = '/content/hw1_workspace'

def setup_data(tar_path, dest_path):
    # Check if the specific data directory (e.g., 'data/train') exists inside the workspace
    if not os.path.exists(os.path.join(dest_path, 'data', 'train')):
        if os.path.exists(dest_path):
            # If dest_path exists but the expected data directory doesn't, remove the existing workspace
            print(f"Incomplete or missing data in {dest_path}. Removing existing workspace...")
            shutil.rmtree(dest_path)

        os.makedirs(dest_path, exist_ok=True) # Ensure dest_path exists
        print("Unpacking data archive...")
        with tarfile.open(tar_path, 'r') as file_ref:
            file_ref.extractall(dest_path)
        print("Unpacking successful!")
    else:
        print("Data is already present in the workspace.")

# Corrected call to setup_data
setup_data(os.path.join(ARCHIVE_LOC, 'cv_hw1_data.tar'), WORKSPACE)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, models, transforms
from PIL import Image

BATCH_SIZE = 256
WORKER_THREADS = 4

# Using RandAugment instead of TrivialAugment/MixUp
training_transformations = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.RandAugment(num_ops=2, magnitude=9),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

testing_transformations = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Reverting paths to use WORKSPACE
path_train = os.path.join(WORKSPACE, 'data', 'train')
path_val = os.path.join(WORKSPACE, 'data', 'val')

ds_train = datasets.ImageFolder(path_train, transform=training_transformations)
ds_val = datasets.ImageFolder(path_val, transform=testing_transformations)

loader_train = DataLoader(ds_train, batch_size=BATCH_SIZE, shuffle=True, num_workers=WORKER_THREADS, pin_memory=True)
loader_val = DataLoader(ds_val, batch_size=BATCH_SIZE, shuffle=False, num_workers=WORKER_THREADS, pin_memory=True)

print(f"Found {len(ds_train)} train and {len(ds_val)} val samples.")
```

```python
COMPUTE_NODE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Custom Pooling: Combines both Max and Average features for richer representation
class ConcatPool2d(nn.Module):
    def __init__(self):
        super().__init__()
        self.ap = nn.AdaptiveAvgPool2d((1, 1))
        self.mp = nn.AdaptiveMaxPool2d((1, 1))
    def forward(self, x):
        return torch.cat([self.mp(x), self.ap(x)], 1)

def build_custom_resnet50():
    # Use ResNet-50 (Different from your ResNet-152)
    net = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)

    # Replace the standard pool with our custom ConcatPool
    net.avgpool = ConcatPool2d()

    # Because ConcatPool doubles the feature maps (Max + Avg), we multiply by 2
    in_features = net.fc.in_features * 2

    net.fc = nn.Sequential(
        nn.Dropout(0.3),
        nn.Linear(in_features, 100)
    )
    return net.to(COMPUTE_NODE)

classifier = build_custom_resnet50()

# SGD with Nesterov Momentum is structurally very different from AdamW
loss_function = nn.CrossEntropyLoss(label_smoothing=0.1)
opt = optim.SGD(classifier.parameters(), lr=0.01, momentum=0.9, weight_decay=1e-4, nesterov=True)

MAX_EPOCHS = 35
# CosineWarmRestarts is a different, highly effective scheduler
lr_scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(opt, T_0=10, T_mult=2)
```

```python
scaler = torch.amp.GradScaler('cuda')

def train_engine(model, dataloader, optimizer, criterion):
    model.train()
    total_loss = 0.0
    for imgs, targets in dataloader:
        imgs, targets = imgs.to(COMPUTE_NODE), targets.to(COMPUTE_NODE)
        optimizer.zero_grad()
        with torch.amp.autocast('cuda'):
            preds = model(imgs)
            loss = criterion(preds, targets)
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        total_loss += loss.item() * imgs.size(0)
    return total_loss / len(dataloader.dataset)

def eval_engine(model, dataloader):
    model.eval()
    correct_preds = 0
    with torch.no_grad():
        for imgs, targets in dataloader:
            imgs, targets = imgs.to(COMPUTE_NODE), targets.to(COMPUTE_NODE)
            preds = model(imgs)
            _, top_class = preds.max(1)
            correct_preds += (top_class == targets).sum().item()
    return correct_preds / len(dataloader.dataset)

# The Main Loop
highest_accuracy = 0.0

for ep in range(MAX_EPOCHS):
    epoch_loss = train_engine(classifier, loader_train, opt, loss_function)
    lr_scheduler.step()
    epoch_acc = eval_engine(classifier, loader_val)

    print(f"Epoch [{ep+1}/{MAX_EPOCHS}] | Loss: {epoch_loss:.4f} | Accuracy: {epoch_acc:.4f}")

    if epoch_acc > highest_accuracy:
        highest_accuracy = epoch_acc
        torch.save(classifier.state_dict(), 'friend_resnet50_best.pth')
        print("   -> New record! Model saved.")

print(f"Done. Peak Accuracy: {highest_accuracy:.4f}")
```

```python
import pandas as pd
from torch.utils.data import Dataset

class UnlabeledDataset(Dataset):
    def __init__(self, folder_path, transform_pipeline):
        self.folder_path = folder_path
        self.file_list = sorted(os.listdir(folder_path))
        self.transform_pipeline = transform_pipeline

    def __len__(self): return len(self.file_list)

    def __getitem__(self, i):
        f_name = self.file_list[i]
        img = Image.open(os.path.join(self.folder_path, f_name)).convert('RGB')
        return self.transform_pipeline(img), f_name

# Load her specific model
classifier.load_state_dict(torch.load('friend_resnet50_best.pth'))
classifier.eval()

path_test = os.path.join(WORKSPACE, 'data', 'test')
ds_test = UnlabeledDataset(path_test, testing_transformations)
loader_test = DataLoader(ds_test, batch_size=BATCH_SIZE, shuffle=False, num_workers=WORKER_THREADS)

# The mandatory folder mapping fix
real_class_names = ds_train.classes
final_predictions = {}

print("Generating final predictions...")
with torch.no_grad():
    for imgs, f_names in loader_test:
        imgs = imgs.to(COMPUTE_NODE)
        with torch.amp.autocast('cuda'):
            # Simple inference without TTA to keep the code distinct
            logits = classifier(imgs)
        _, winning_idx = torch.max(logits, 1)

        for name, idx in zip(f_names, winning_idx):
            clean_string = name.replace('.jpg', '')
            actual_folder = int(real_class_names[idx.item()])
            final_predictions[clean_string] = actual_folder

# Format for CodaBench
submission_template = pd.read_csv('/content/sample prediction.csv')
submission_template['pred_label'] = submission_template['image_name'].map(final_predictions).fillna(submission_template['pred_label']).astype(int)
submission_template.to_csv('prediction.csv', index=False)
print("Saved to prediction.csv successfully.")
```
